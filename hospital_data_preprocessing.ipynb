# 데이터 로드

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original = pd.read_csv('/gdrive/MyDrive/Petom/csv_files/행정안전부_동물병원.csv', encoding='cp949')
original.head()
```

```python
original.info()
```

## 컬럼별 null값 개수 확인

```python
original.isnull().sum()
```

## '영업상태명' count 확인

```python
original.groupby('영업상태명').count()
```

# '영업상태명' 컬럼에서 '영업/정상'만 뽑기

```python
original_open = original[original['영업상태명']=='영업/정상']
original_open.head()
```

```python
# null값 개수 확인

original_open.isnull().sum()
```

# 좌표정보가 null인 Row 삭제

```python
hospital_loc = original_open.dropna(subset=['좌표정보(x)']).reset_index()
hospital_loc.head()
```

# 중부원점TM(EPSG:2097)을 위도경도(WGS84)로 변환

```python
from pyproj import Proj, transform
proj_1 = Proj(init='epsg:2097')
proj_2 = Proj(init='epsg:4326')

converted = transform(proj_1, proj_2, hospital_loc['좌표정보(x)'].values, hospital_loc['좌표정보(y)'].values)
hospital_loc['경도'] = converted[0]
hospital_loc['위도'] = converted[1]
```

# 필요한 컬럼만 뽑기

```python
hospital = hospital_loc[['상세영업상태명', '소재지전화','소재지전체주소','도로명전체주소','도로명우편번호','사업장명','경도','위도']]
hospital.head()
```

```python
hospital.info()
```

# 데이터 전처리 최종 확인

## 컬럼별 null값 확인

```python
hospital.isnull().sum()
```

## '도로명우편주소' 컬럼의 데이터 형식 str으로 변경  

```python
# '도로명우편주소'에 포함된 null값 0으로 채우기
# -1을 'none'으로 변경
hospital['도로명우편번호'] = hospital['도로명우편번호'].fillna(0)

# '도로명우편주소' datatype 을 float -> int로 변경
hospital['도로명우편번호'] = hospital['도로명우편번호'].astype(int)

# '도로명우편주소' datatype 을 int -> str로 변경
hospital['도로명우편번호'] = hospital['도로명우편번호'].astype(str)

# '도로명우편주소' == '0'을 'none'으로 변경
hospital['도로명우편번호'] = hospital['도로명우편번호'].replace({'0': None})

hospital.head()
```
